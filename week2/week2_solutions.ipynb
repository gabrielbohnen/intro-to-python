# **Challenge Question 1**
Write a Python program to create a class representing a Cylinder. Include methods to calculate its surface area and volume.

```python
import math

PI = math.pi

class Cylinder:
    
  def __init__(self, radius, height):
    # initalise the radius and height of the Cylinder object (Hint: you will need to use self.radius)
    self.radius = radius
    self.height = height

  def surface_area(self):
    # Calculate and return the surface area of the Cylinder 
    return 2*PI*self.radius**2+2*PI*self.radius*self.height
    
  # Define a function to calculate volume called 'volume'
  # Within the function calculate and return the volume of the Cylinder
  def volume(self):
    return PI*self.radius**2*self.height
```

```python
# Test your code here
cyl1 = Cylinder(5,10) # Cylinder with radius 5 and height 10

surface_area = cyl1.surface_area()
volume = cyl1.volume()

print(f"A cylinder with radius {cyl1.radius} and height {cyl1.height} has a surface area of {surface_area} and volume {volume}")
```

# **Challlenge Question 2**

Implement a simple class-based system to manage books in a library.

#### Tasks

1. **Create a `Book` class**
    - The class should have attributes for `title`, `author`, and `year`.
    - Add a method called `display_info` to display information about the book.

2. **Create a `Library` class**
    - The class should maintain a collection (as a list) of `Book` objects.
    - Add a method called `add_book` to add a book to the library if it is not already in the library.
    - Add a method called `remove_book` that takes a `title` as input and removes a book based on the `title` 
    - Add a method called `display_books` that displays the info for each book in the library

### Instructions

- Use constructors (`__init__`), instance methods, and object composition.
- Make sure to use encapsulation to keep the data safe within each class.
- Print outputs to demonstrate each methodâ€™s functionality.


```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def display_info(self):
        return f"{self.title} by {self.author} ({self.year})"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, title):
        for b in self.books:
            if b.title == title:
                self.books.remove(b)
                return True
        return False

    def display_books(self):
        return [book.display_info() for book in self.books]
```

### Demonstrate usage
Create at least three books and add them to the library. Show how to remove a book and display the remaining books.

```python
library = Library()
book1 = Book("1984", "George Orwell", 1949)
book2 = Book("To Kill a Mockingbird", "Harper Lee", 1960)
book3 = Book("The Great Gatsby", "F. Scott Fitzgerald", 1925)

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

print("All books:", library.display_books())

library.remove_book("1984")

print("After removal:", library.display_books())
```

1. Define a function to calculate the factorial of a number.

```python
# factorial function uses recursion - recursion is when a function calls itself
def factorial(n):
    # Calculate the factorial of n
    if n == 0 or n == 1: ## <--- FIX: Colon added in ##
        return 1
    else:
        return n * factorial(n - 1) ## <--- FIX: Factorial is multiplication, not addition ##

def is_palindrome(word):
    # Check if a word is a palindrome
    return word == word[::-1] ## <--- FIX: We need to compare the word to its reverse form, adding in the negative sign achieves this  ##
```

```python
# Test out your code here

factorial_output = factorial(3)
print(f"3 Factorial should equal 6, the output from the function is: {factorial_output}")

palindrome_word = "tacocat"
non_palindrome_word = "cheese"

print(f"Is {palindrome_word} a palindrome? {is_palindrome(palindrome_word)}")
print(f"Is {non_palindrome_word} a palindrome? {is_palindrome(non_palindrome_word)}")
```

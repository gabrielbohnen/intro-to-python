```python
import pandas as pd
import numpy as np

# Number of rows
n_rows = 100_000

# Example lists for categorical columns
transformers = [f'Transformer_{str(i).zfill(3)}' for i in range(1, 51)]
feeders = [f'Feeder_{str(i).zfill(3)}' for i in range(1, 21)]
suburbs = ['Brendale', 'Strathpine', 'Petrie', 'Eatons Hill', 'Albany Creek']
ev_models = ['Tesla Model 3', 'Nissan Leaf', 'Hyundai Ioniq']

transformers_feeders = pd.DataFrame({
    'Transformer': transformers,
    'Feeder': np.random.choice(feeders, 50)
})

feeders_suburbs = pd.DataFrame({
    'Feeder': feeders,
    'Suburb': np.random.choice(suburbs, 20)
})

# Randomly generate data
challenge_3_df = pd.DataFrame({
    'NMI': [f'NMI{str(i).zfill(6)}' for i in range(n_rows)],
    'Transformer': np.random.choice(transformers, n_rows),
    'InstallationDate': pd.to_datetime(np.random.choice(pd.date_range('2010-01-01', '2024-12-31'), n_rows)),
    # Initialize as object for missing values
    'HasSolar': np.random.choice([True, False], n_rows).astype(object),
    'HasEV': np.random.choice([True, False], n_rows).astype(object),
    'HasBESS': np.random.choice([True, False], n_rows).astype(object),
})

# Assign SolarSize only if HasSolar is True, else 0
challenge_3_df['SolarSize'] = np.where(
    challenge_3_df['HasSolar'] == True,
    np.round(np.random.uniform(2, 10, n_rows), 1),
    0
)

# Assign EVmodel only if HasEV is True, else None
challenge_3_df['EVmodel'] = np.where(
    challenge_3_df['HasEV'] == True,
    np.random.choice(ev_models, n_rows),
    "No EV"
)

# Assign BatterySize only if HasBESS is True, else 0
challenge_3_df['BatterySize'] = np.where(
    challenge_3_df['HasBESS'] == True,
    np.round(np.random.uniform(5, 20, n_rows), 1),
    0
)

# Optionally insert more missing values for practice
for col in ['NMI','Transformer','InstallationDate', 'HasSolar', 'HasEV', 'HasBESS', 'SolarSize', 'EVmodel', 'BatterySize']:
    missing_rate = np.random.uniform(0.03,0.07) 
    mask = np.random.rand(n_rows) < missing_rate  # between 3% to 7% missing values
    challenge_3_df.loc[mask, col] = None
```
